1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
def convert(number): 
# WE CONSIDER INPUT number AS A STRING OF CHARACTHERS, NOT A NUMBER with type INT
# we also output as a string
# only positive number


    # if input is in binary (starts with '0b')
    if number[:2]=='0b':    
        number = number[2:] # we remove the '0b' prefix

        # conversion to decimal : 
        dec = 0
        for i in range(len(number)):
            dec = dec + int(number[-(i+1)])*2**i
        dec = str(dec)
        print("decimal : ", dec)


        # conversion to hexadecimal (using the decimal form just found) :
        hex=''
        dec = int(dec)
        a = dec
        b = 16

        if dec == 0:
            hex = '0x0'
        else : 
            while a != 0:
                q, r = divmod(a, b)
                r = str(r)
                r = r.replace("10","A").replace("11","B").replace("12","C").replace("13","D").replace("14","E").replace("15","F")
                hex += r
                a = q
            hex += "x0"
            hex = hex[::-1] #[::-1] take the entire string from start to end but in reverse order, stepping backwards with -1

        print("hex : ", hex)

        dec = str(dec)
        return dec, hex
    

    # if input is in hexadecimal (starts with '0x')
    elif number[:2]=='0x':  
        number = number[2:] # we remove the '0x' prefix

        # conversion to decimal :
        dec = 0
        for i in range(len(number)):
            #i
            num = number[-(i+1)].upper().replace("A","10").replace("B","11").replace("C","12").replace("D","13").replace("E","14").replace("F","15")
            dec = dec + int(num)*16**i
        dec = str(dec)
        print("decimal : ", dec)

        # conversion to binary (using the decimal form just found) :
        bin = ''
        dec = int(dec)
        a = dec
        b = 2

        if dec == 0:
            bin = '0b0'
        else : 
            while a != 0:
                q, r = divmod(a, b)
                r = str(r)
                bin += r
                a = q
            bin += 'b0'
            bin = bin[::-1] #[::-1] take the entire string from start to end but in reverse order, stepping backwards with -1

        print("binary", bin)

        dec = str(dec)
        return dec, bin



    # else input is in decimal
    else :  

        # conversion to binary :
        bin = ''
        dec = int(number)
        a = dec
        b = 2

        if dec == 0:
            bin = '0b0'
        else:
            while a != 0:
                q, r = divmod(a, b)
                r = str(r)
                bin += r
                a = q
            bin += 'b0'
            bin = bin[::-1] #[::-1] take the entire string from start to end but in reverse order, stepping backwards with -1

        print("binary", bin)

        # conversion to hexadecimal :
        hex=''
        dec = int(number)
        a = dec
        b = 16

        if dec == 0:
            hex = '0x0'
        else:
            while a != 0:
                q, r = divmod(a, b)
                r = str(r)
                r = r.replace("10","A").replace("11","B").replace("12","C").replace("13","D").replace("14","E").replace("15","F")
                hex += r
                a = q
            hex += "x0"
            hex = hex[::-1] #[::-1] take the entire string from start to end but in reverse order, stepping backwards with -1

        print("hex : ", hex)


        return bin, hex
```
